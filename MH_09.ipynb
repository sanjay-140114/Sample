```python
############################### IMP PROMPT EXECUTION ####################

response = query_engine.query(overall_royalty_extraction_prompt)
print(str(response))

response_str = str(response)
str_to_replace = {'%': '','Task-1': 'Domestic Sales', 'Task-2': 'Foreign', 'Task-3': 'Canadian','Task-4': 'High Discount 1', 'Task-5': 'State Adoptions', 'Task-6': 'Chapter Sales','Task-7': 'Subscription Sales - Domestic', 'Task-8': 'Subscription Sales - foreign', 'Task-9': 'Subscription - Trial/present', 'Task-m': 'Author ref file id', 'Task-n': 'Author ref id'}

# Iterate over all key-value pairs in dictionary
for key, value in str_to_replace.items():
    # Replace key character with value character in string
    response_str = response_str.replace(key, value)

print(response_str)

response_json = json.loads(response_str)
#print(response_json)

```

```python
Domestic_Sales_v                          = bool(response_json['Domestic Sales'].replace(".", "").isnumeric())
Foreign_v                                 = bool(response_json['Foreign'].replace(".", "").isnumeric())
Canadian_v                                = bool(response_json['Canadian'].replace(".", "").isnumeric())
High_Discount_1_v                         = bool(response_json['High Discount 1'].replace(".", "").isnumeric())
State_Adoptions_v                         = bool(response_json['State Adoptions'].replace(".", "").isnumeric())
Chapter_Sales_v                           = bool(response_json['Chapter Sales'].replace(".", "").isnumeric())

SS_Domestic_v                             = bool(response_json['Subscription Sales - Domestic'].replace(".", "").isnumeric())
SS_foreign_v                              = bool(response_json['Subscription Sales - foreign'].replace(".", "").isnumeric())
SS_Trial_v                                = bool(response_json['Subscription - Trial/present'].replace(".", "").isnumeric())
Author_ref_file_v                         = bool(response_json['Author ref file id'].replace(".", "").isnumeric())
Author_ref_v                              = bool(response_json['Author ref id'].replace(".", "").isnumeric())
```
